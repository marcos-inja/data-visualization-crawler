```python
from bs4 import BeautifulSoup
import requests
# Futuramente
# import concurrent.futures

BASE_URL = "https://myanimelist.net/profile"

# Pega a pagina principal do usuario
def get_page_user(url_part_name):
    url_format = f'{BASE_URL}/{url_part_name}'

    page_source = requests.get(url_format)
    soup = BeautifulSoup(page_source.text, 'html.parser')
    
    return soup


# Pega a pagina de amigos do usuario
def get_page_frinds(url_part_name):
    url_format = f'{BASE_URL}/{url_part_name}/friends'

    page_source = requests.get(url_format)
    soup = BeautifulSoup(page_source.text, 'html.parser')
    
    return soup


# Pega as informações do usuario
def user_info(soup):
    user_status = soup.find(class_ = 'user-status').find_all("li")
    info = []
    for i in user_status:
        tag = i.find_all(text=True)
        info.append(tag)
        
    return info


# Pega os amigos do usuario
def user_friends(soup):
    info = []
    try:
        user_status = soup.find(class_ = 'friend').find_all(class_ = 'boxlist')
        for i in user_status:
            tag = i.find(class_ = 'title').find('a')
            name = tag.text
            info.append(name)
    except:
        pass
    return info


# Pega as informações dos amigos do usuario
def friends_data(friends):
    data =[]
    if friends:
        for f in friends:
            soup = get_page_user(f)
            infos_user = user_info(soup)

            soup = get_page_frinds(f)
            friends = user_friends(soup)
            data.append([infos_user, friends])
            # print(f"{f}:\nInfo: {infos_user}\nAmigos:{friends}")
    return data


def main(url_part_name):
    soup = get_page_user(url_part_name)
    # User info
    infos_user = user_info(soup)
    # User friends
    soup = get_page_frinds(url_part_name)
    friends = user_friends(soup)
    # Pegando dados dos amigos do perfil principal
    foff = friends_data(friends)
    for f in foff:
        print(f'{f}\n\n')
    
    
main("khalil04uzumaki")
```
